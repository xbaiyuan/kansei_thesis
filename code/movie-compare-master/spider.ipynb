```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.selector import Selector

class RestaurantSpider(scrapy.Spider):
    name = "restaurants"
    start_urls = [base_url+ 'p'+str(page) for page in range(1,51)]
#     start_urls = [
#         'http://www.dianping.com/search/category/1/10/o3',
#     ]

    def parse(self, response):
        selector = Selector(response)
        div = selector.xpath('//div[@id="shop-all-list"]/ul/li')
        for dd in div:
            title = dd.xpath('div[2]/div[1]/a[1]/h4/text()').extract_first()
            image_url = dd.css('.pic img::attr(data-src)').extract_first()
            shopurls = dd.xpath('div[2]/div[1]/a[1]/@href').extract()
            url = 'http://www.dianping.com'+str(shopurls[0])

            branch = dd.xpath('div[2]/div[1]/a[2]/text()').extract_first()

            ratings_num = int(dd.xpath('div[2]/div[2]/a[1]/b/text()').extract_first())
            price = dd.xpath('div[2]/div[2]/a[2]/b/text()').extract_first()

            taste = float(dd.xpath('div[2]/span/span[1]/b/text()').extract_first())
            environment = float(dd.xpath('div[2]/span/span[2]/b/text()').extract_first())
            service = float(dd.xpath('div[2]/span/span[3]/b/text()').extract_first())

            foodtype = dd.xpath('div[2]/div[3]/a[1]/span/text()').extract_first()
            location = dd.xpath('div[2]/div[3]/a[2]/span/text()').extract_first()
            address = dd.xpath('div[2]/div[3]/span[1]/text()').extract_first()

            yield {
                'title': title,
                'image_url': image_url,
                'url': url,
                'ratings_num': ratings_num,
                'branch': branch, 
                'price': price,
                'taste': taste,
                'environment': environment,
                'service': service,
                'type': foodtype,
                'location': location,
                'address': address
            }
```

```python
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)',
    'FEED_FORMAT': 'json',
    'FEED_URI': 'jiading.json',
    'FEED_EXPORT_ENCODING':'utf-8',
    'COOKIES_ENABLED': False,
    'DOWNLOAD_DELAY': 3,
})

process.crawl(RestaurantSpider)
process.start() 
```

```python
file_name = './data/cn/2016_movie.csv'
df = pd.read_csv(file_name, header = 0)
movie_links = df['link'].tolist()
```

```python
import scrapy
from scrapy.crawler import CrawlerProcess

class MovieSpider(scrapy.Spider):
    name = "movies"
    start_urls = movie_links
#     start_urls = [
#         'https://movie.douban.com/subject/26388624/',
#         'https://movie.douban.com/subject/25934014/',
#         'https://movie.douban.com/subject/25852314/',
#         'https://movie.douban.com/subject/25858763/',
#         'https://movie.douban.com/subject/4811813/'
#     ]

    def parse(self, response):
        title = response.css('div#wrapper div#content h1 span::text').extract_first()
        year = response.css('div#wrapper div#content h1 span.year::text').extract_first().strip('()')
        img_url = response.css('div#mainpic a.nbgnbg img::attr(src)').extract_first()
        
        rating = response.css('div.rating_self.clearfix strong.ll.rating_num::text').extract_first()
        star_distributions = response.css('div.ratings-on-weight div.item span.rating_per::text').extract()
        ratings_num = response.css('div.rating_sum a.rating_people span::text').extract_first()
        
        imdb_id = ''
        for link in response.css('div#info a'):
            if 'imdb' in link.css('::attr(href)').extract_first():
                imdb_id = link.css('::text').extract_first()

        yield {
            'url': response.url,
            'title': title,
            'img_url': img_url,
            'year': year,
            'rating': rating,
            'ratings_num': ratings_num,
            '5_star': star_distributions[0],
            '4_star': star_distributions[1],
            '3_star': star_distributions[2],
            '2_star': star_distributions[3],
            '1_star': star_distributions[4],
            'imdb_id': imdb_id
        }
```

```python
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)',
    'FEED_FORMAT': 'json',
    'FEED_URI': 'movie_2016.json',
    'FEED_EXPORT_ENCODING':'utf-8',
    'COOKIES_ENABLED': False,
    'DOWNLOAD_DELAY': 2,
})

process.crawl(MovieSpider)
process.start() 
```
